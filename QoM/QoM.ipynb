### Removing excess columns

```python
def trim_df(participant):
    for df in participant:
        # Remove unwanted markers for QoM calculation
        for i in df.columns:
            # Remove static piano and clap marker
            if i == 'Piano_LF' or i == 'Piano_LB' or i == 'Piano_RF' or i == 'Piano_RB' or i == 'Hollywood':
                df.drop(i, inplace=True, axis=1)
            # Remove wrist markers
            elif i == 'L_Wrist_L' or i == 'L_Wrist_R' or i == 'R_Wrist_L' or i == 'R_Wrist_R':
                df.drop(i, inplace=True, axis=1)
        # Remove non-marker columns (index, timestamp, newtime, task)
        for i in df.columns[-4:]:
            df.drop(i, inplace=True, axis=1)
```

```python
trim_df(P1)
trim_df(P2)
trim_df(P3)
trim_df(P4)
```

### Making a dictionary of participants/tasks

```python
dataframes = {}

for p in range(4):
    participant = 'P' + str(p + 1)
    dataframes[participant] = {}
    
for t in range(12):
    task = 'task' + str(t + 1)
    dataframes['P1'][task] = P1[t]
    dataframes['P2'][task] = P2[t]
    dataframes['P3'][task] = P3[t]
    dataframes['P4'][task] = P4[t]
```

```python
# Making a dictionary of QoM for each marker of each task of each participant:
qom_dict = {}

for prt in dataframes:
    # Participant
    qom_dict[prt] = {}
    for tsk in dataframes[prt]:
        # Task
        qom_dict[prt][tsk] = {}
        for mrk in dataframes[prt][tsk]:
            # Marker
            qom_dict[prt][tsk][mrk] = {}
            # Mean and Standard deviation
            qom_dict[prt][tsk][mrk][mrk+'_mean'] = dataframes[prt][tsk][mrk].mean()
            qom_dict[prt][tsk][mrk][mrk+'_std'] = dataframes[prt][tsk][mrk].std()
        
```

```python
qom_dict['P4']['task12']['Head']
```

## Defining cases


**Case 1**: All markers (except wrist markers)

    Too much motion centered around hands with both hand and double wrist markers (3 markers pr hand)
    
**Case 2**: Except head marker

    To plot the combined markers of the body vs only head marker
    
**Case 3**: Only head marker

```python
qom_cases = {}

# CASE 1
case = 'case1'
qom_cases[case] = {}

# Participant
for prt in dataframes:
    qom_cases[case][prt] = {}
    # Task
    for tsk in dataframes[prt]:
        qom_cases[case][prt][tsk] = {}
        qom_cases[case][prt][tsk][tsk+'_mean'] = []
        qom_cases[case][prt][tsk][tsk+'_std'] = []
        tsk_tmp = 0
        
        # Marker
        for mrk in dataframes[prt][tsk]:            
            # Listing means
            qom_cases[case][prt][tsk][tsk+'_mean'].append(qom_dict[prt][tsk][mrk][mrk+'_mean'])         
        # Mean & std calculation
        qom_cases[case][prt][tsk][tsk+'_std'] = np.std(qom_cases[case][prt][tsk][tsk+'_mean'])
        qom_cases[case][prt][tsk][tsk+'_mean'] = np.mean(qom_cases[case][prt][tsk][tsk+'_mean'])
        
```

```python
# CASE 2
case = 'case2'
qom_cases[case] = {}

# Participant
for prt in dataframes:
    qom_cases[case][prt] = {}
    # Task
    for tsk in dataframes[prt]:
        qom_cases[case][prt][tsk] = {}
        qom_cases[case][prt][tsk][tsk+'_mean'] = []
        qom_cases[case][prt][tsk][tsk+'_std'] = []
        tsk_tmp = 0
        
        # Marker
        for mrk in dataframes[prt][tsk]:
            if mrk != 'Head':
                # Listing means
                qom_cases[case][prt][tsk][tsk+'_mean'].append(qom_dict[prt][tsk][mrk][mrk+'_mean'])         
        # Mean & std calculation
        qom_cases[case][prt][tsk][tsk+'_std'] = np.std(qom_cases[case][prt][tsk][tsk+'_mean'])
        qom_cases[case][prt][tsk][tsk+'_mean'] = np.mean(qom_cases[case][prt][tsk][tsk+'_mean'])
        
```

```python
# CASE 3
case = 'case3'
qom_cases[case] = {}

# Participant
for prt in dataframes:
    qom_cases[case][prt] = {}
    # Task
    for tsk in dataframes[prt]:
        qom_cases[case][prt][tsk] = {}
        qom_cases[case][prt][tsk][tsk+'_mean'] = []
        qom_cases[case][prt][tsk][tsk+'_std'] = []
        tsk_tmp = 0
        
        # Marker
        for mrk in dataframes[prt][tsk]:
            if mrk == 'Head':
                # Listing means & stds
                qom_cases[case][prt][tsk][tsk+'_mean'].append(qom_dict[prt][tsk][mrk][mrk+'_mean'])
                qom_cases[case][prt][tsk][tsk+'_std'].append(qom_dict[prt][tsk][mrk][mrk+'_std'])                
        # Mean & std calculation
        qom_cases[case][prt][tsk][tsk+'_mean'] = np.mean(qom_cases[case][prt][tsk][tsk+'_mean'])
        qom_cases[case][prt][tsk][tsk+'_std'] = qom_cases[case][prt][tsk][tsk+'_std'][0]  
      
```
