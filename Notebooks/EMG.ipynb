### Removing excess columns

```python
def trim_df(participant):
    for df in participant:
        # Remove unwanted columns for EMG calculation
        for i in df.columns[:2]:
            df.drop(i, inplace=True, axis=1)
        for i in df.columns[-2:]:
            df.drop(i, inplace=True, axis=1)
```

```python
trim_df(P1)
trim_df(P2)
trim_df(P3)
trim_df(P4)
```

### Making a dictionary of participants/tasks

```python
dataframes = {}

for p in range(4):
    participant = 'P' + str(p + 1)
    dataframes[participant] = {}
    
for t in range(12):
    task = 'task' + str(t + 1)
    dataframes['P1'][task] = P1[t]
    dataframes['P2'][task] = P2[t]
    dataframes['P3'][task] = P3[t]
    dataframes['P4'][task] = P4[t]
```

# Average EMG

```python
dataframes['P4']['task1'].mean()
```

```python
dataframes['P4']['task1'].std(ddof=0)
```

```python
# Making a dictionary of average EMG for each sensor of each task of each participant:
emg_dict = {}

for prt in dataframes:
    # Participant
    emg_dict[prt] = {}
    for tsk in dataframes[prt]:
        # Task
        emg_dict[prt][tsk] = {}
        for sns in dataframes[prt][tsk]:
            # Sensor
            emg_dict[prt][tsk][sns] = {}
            # Mean and Standard deviation
            emg_dict[prt][tsk][sns][sns+'_mean'] = dataframes[prt][tsk][sns].mean()
            emg_dict[prt][tsk][sns][sns+'_std'] = dataframes[prt][tsk][sns].std(ddof=0)
        
```

```python
emg_dict['P4']['task1']
```

## Defining cases


**Case 1**: All sensors

    Average of both flexor and extensor of each hand
    
**Case 2**: Flexors (Both hands)

    Average of flexors of both hands
    
**Case 3**: Extensors (Both hands)

    Average of extensors of both hands 
    
**Case 4**: Left hand extensor

    Average of Left hand extensor
    
**Case 5**: Right hand extensor

    Average of Right hand extensor (EMG data of choice)

```python
emg_cases = {}

# CASE 1
case = 'case1'
emg_cases[case] = {}

# Participant
for prt in dataframes:
    emg_cases[case][prt] = {}
    # Task
    for tsk in dataframes[prt]:
        emg_cases[case][prt][tsk] = {}
        emg_cases[case][prt][tsk][tsk+'_mean'] = []
        emg_cases[case][prt][tsk][tsk+'_std'] = []
        
        # Marker
        for sns in dataframes[prt][tsk]:            
            # Listing means
            emg_cases[case][prt][tsk][tsk+'_mean'].append(emg_dict[prt][tsk][sns][sns+'_mean'])         
        # Mean & std calculation
        emg_cases[case][prt][tsk][tsk+'_std'] = np.nanstd(emg_cases[case][prt][tsk][tsk+'_mean'])
        emg_cases[case][prt][tsk][tsk+'_mean'] = np.nanmean(emg_cases[case][prt][tsk][tsk+'_mean'])    
```

```python
# CASE 2
case = 'case2'
emg_cases[case] = {}

# Participant
for prt in dataframes:
    emg_cases[case][prt] = {}
    # Task
    for tsk in dataframes[prt]:
        emg_cases[case][prt][tsk] = {}
        emg_cases[case][prt][tsk][tsk+'_mean'] = []
        emg_cases[case][prt][tsk][tsk+'_std'] = []
        
        # Marker
        for sns in dataframes[prt][tsk]:
            if sns == 'left_flexor' or sns == 'right_flexor':
                # Listing means
                emg_cases[case][prt][tsk][tsk+'_mean'].append(emg_dict[prt][tsk][sns][sns+'_mean'])         
        # Mean & std calculation
        emg_cases[case][prt][tsk][tsk+'_std'] = np.nanstd(emg_cases[case][prt][tsk][tsk+'_mean'])
        emg_cases[case][prt][tsk][tsk+'_mean'] = np.nanmean(emg_cases[case][prt][tsk][tsk+'_mean'])
```
