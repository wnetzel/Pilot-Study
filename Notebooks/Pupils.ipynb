### Removing excess columns

```python
def trim_df(participant):
    for df in participant:
        # Remove unwanted markers for pupil calculation
        for i in df.columns:
            # Use only 'xmm_filled'
            if i != 'xmm_filled':
                df.drop(i, inplace=True, axis=1)
```

```python
trim_df(P1)
trim_df(P2)
trim_df(P3)
trim_df(P4)
```

### Making a dictionary of participants/tasks

```python
dataframes = {}

for p in range(4):
    participant = 'P' + str(p + 1)
    dataframes[participant] = {}
    
for t in range(12):
    task = 'task' + str(t + 1)
    dataframes['P1'][task] = P1[t]
    dataframes['P2'][task] = P2[t]
    dataframes['P3'][task] = P3[t]
    dataframes['P4'][task] = P4[t]
```

```python
dataframes['P4']['task1']
```

# Average Pupil Size

```python
dataframes['P4']['task1'].mean()
```

```python
dataframes['P4']['task1'].std()
```

```python
# Making a dictionary of Average Pupil Size for task of each participant:
avg_pup_dict = {}

for prt in dataframes:
    # Participant
    avg_pup_dict[prt] = {}
    for tsk in dataframes[prt]:
        # Task
        avg_pup_dict[prt][tsk] = {}
        avg_pup_dict[prt][tsk][tsk+'_mean'] = dataframes[prt][tsk].mean()[0]
        avg_pup_dict[prt][tsk][tsk+'_std'] = dataframes[prt][tsk].std()[0]        


        
```

```python
avg_pup_dict['P4']['task1']
```

## Defining cases


**Case 1**: All markers (except wrist markers)

    Too much motion centered around hands with both hand and double wrist markers (3 markers pr hand)
    
**Case 2**: Except head marker

    To plot the combined markers of the body vs only head marker
    
**Case 3**: Only head marker

```python
# Printing the mean of the pupil size of each task pr participant
print('\nAverage Pupil Size (mm)')
for prt in avg_pup_dict:
    # Participant
    print(f'\n{prt}\t{"": <3}Mean{"": <5}\tStd\n')
    for tsk in avg_pup_dict[prt]:
        # The mean rounded to 5 decimals
        print(tsk, f'\t{"": <3}{avg_pup_dict[prt][tsk][tsk+"_mean"]:.5}{"": <3}\t{avg_pup_dict[prt][tsk][tsk+"_std"]:.5}')
    print('\n')
```

## Plotting the Average Pupil Size of each task pr participant

```python
# Making a list of labels
labels = []

# Task labels (similar for all participants)
for tsk in avg_pup_dict['P1']:
    labels.append(tsk)
```

### Making lists for plotting bars and errorbars

```python
# Making a list of Pul Size means & std
avg_pup_plot = []
avg_pup_yerr = []


# Participant
for i, prt in enumerate(avg_pup_dict):
    avg_pup_plot.append([])
    avg_pup_yerr.append([])
    # Task
    for tsk in avg_pup_dict[prt]:
        avg_pup_plot[i].append(avg_pup_dict[prt][tsk][tsk+'_mean'])
        avg_pup_yerr[i].append(avg_pup_dict[prt][tsk][tsk+'_std'])
```

```python
fig, axs = plt.subplots(2, 2, figsize=(24, 12), sharey=True)

fig.suptitle('Average Pupil Size', fontsize=32)

rects1 = axs[0,0].bar(labels, avg_pup_plot[0], color='c')
rects2 = axs[0,1].bar(labels, avg_pup_plot[1], color='m')
rects3 = axs[1,0].bar(labels, avg_pup_plot[2], color='y')
rects4 = axs[1,1].bar(labels, avg_pup_plot[3], color='g')

axs[0,0].set_title('P1', fontsize=16)
axs[0,1].set_title('P2', fontsize=16)
axs[1,0].set_title('P3', fontsize=16)
axs[1,1].set_title('P4', fontsize=16)

axs[0,0].set_ylim(0)
axs[0,1].set_ylim(0)
axs[1,0].set_ylim(0)
axs[1,1].set_ylim(0)

axs[0,0].bar_label(rects1, label_type='center')
axs[0,1].bar_label(rects2, label_type='center')
axs[1,0].bar_label(rects3, label_type='center')
axs[1,1].bar_label(rects4, label_type='center')

fig.tight_layout()
plt.savefig('./plots/avg_pupil_bar.pdf')
plt.show()
```

```python
fig, axs = plt.subplots(2, 2, figsize=(24, 12), sharey=True)

fig.suptitle('Average Pupil Size (w/errorbars)', fontsize=32)

rects1 = axs[0,0].bar(labels, avg_pup_plot[0], yerr=avg_pup_yerr[0], capsize=6, color='c')
rects2 = axs[0,1].bar(labels, avg_pup_plot[1], yerr=avg_pup_yerr[1], capsize=6, color='m')
rects3 = axs[1,0].bar(labels, avg_pup_plot[2], yerr=avg_pup_yerr[2], capsize=6, color='y')
rects4 = axs[1,1].bar(labels, avg_pup_plot[3], yerr=avg_pup_yerr[3], capsize=6, color='g')

axs[0,0].set_title('P1', fontsize=16)
axs[0,1].set_title('P2', fontsize=16)
axs[1,0].set_title('P3', fontsize=16)
axs[1,1].set_title('P4', fontsize=16)

axs[0,0].set_ylim(0)
axs[0,1].set_ylim(0)
axs[1,0].set_ylim(0)
axs[1,1].set_ylim(0)

axs[0,0].bar_label(rects1, label_type='center')
axs[0,1].bar_label(rects2, label_type='center')
axs[1,0].bar_label(rects3, label_type='center')
axs[1,1].bar_label(rects4, label_type='center')

fig.tight_layout()
plt.savefig('./plots/avg_pupil_errorbar.pdf')
plt.show()
```

## Comparing Participants pr task for each case

```python
x = np.arange(len(labels))  # the label locations
width = 0.4  # the width of the bars

fig, axs = plt.subplots(figsize=(24,6))
rects1 = axs.bar(x - width/2, avg_pup_plot[0], width/2, label='P1')
rects2 = axs.bar(x, avg_pup_plot[1], width/2, label='P2')
rects3 = axs.bar(x + width/2, avg_pup_plot[2], width/2, label='P3')
rects4 = axs.bar(x + width, avg_pup_plot[3], width/2, label='P4')


# Add text for labels, title and custom x-axis tick labels, etc.
axs.set_ylabel('Avg Pupil Size (mm)', fontsize=16)
axs.set_title('Pupil Size by Participant Comparison', pad=16, fontsize=24)
axs.set_xticks(x+0.1, labels)
axs.legend()


# Adding bar labels (mean)
axs.bar_label(rects1, padding=3)
axs.bar_label(rects2, padding=3)
axs.bar_label(rects3, padding=3)
axs.bar_label(rects4, padding=3)


# Tightening, plotting and saving
fig.tight_layout()
plt.savefig('./plots/avg_pupil_participant_comparison_bar.pdf')  
plt.show()
```
