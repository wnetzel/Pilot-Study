```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import joblib

sns.set()
```

```python
path = './plot_data/'

pupillometry = joblib.load(path+'avg_pup_plot_z')
emg = joblib.load(path+'emg_case6_plot')
qom_body = joblib.load(path+'qom_case2_plot')
qom_head = joblib.load(path+'qom_case3_plot')
midi = joblib.load(path+'avg_vel_plot')
cogn_diff = joblib.load(path+'cogn_diff')
tech_diff = joblib.load(path+'tech_diff')
```

```python
P1_corr = {'emg': emg[0],
           'body': qom_body[0],
           'head': qom_head[0],
           'pupil': pupillometry[0],
           'MIDI': midi[0],
           'cogn_diff': cogn_diff['P1_cognitive'],
           'tech_diff': tech_diff['P1_technical']
          }

P1_corr = pd.DataFrame(P1_corr)


P2_corr = {'emg': emg[1],
           'body': qom_body[1],
           'head': qom_head[1],
           'pupil': pupillometry[1],
           'MIDI': midi[1],
           'cogn_diff': cogn_diff['P2_cognitive'],
           'tech_diff': tech_diff['P2_technical']
          }

P2_corr = pd.DataFrame(P2_corr)


P3_corr = {'emg': emg[2],
           'body': qom_body[2],
           'head': qom_head[2],
           'pupil': pupillometry[2],
           'MIDI': midi[2],
           'cogn_diff': cogn_diff['P3_cognitive'],
           'tech_diff': tech_diff['P3_technical']
          }

P3_corr = pd.DataFrame(P3_corr)


P4_corr = {'emg': emg[3],
           'body': qom_body[3],
           'head': qom_head[3],
           'pupil': pupillometry[3],
           'MIDI': midi[3],
           'cogn_diff': cogn_diff['P4_cognitive'],
           'tech_diff': tech_diff['P4_technical']
          }

P4_corr = pd.DataFrame(P4_corr)
```

# Correlation Heatmap (per participant)

```python
# Making rounded correlation values
P1_mat = P1_corr.corr()
round(P1_mat,2)

P2_mat = P2_corr.corr()
round(P2_mat,2)

P3_mat = P3_corr.corr()
round(P3_mat,2)

P4_mat = P4_corr.corr()
round(P4_mat,2)


# Plotting
fig, axs = plt.subplots(2, 2, figsize=(12, 10))
fig.suptitle('Correlation Heatmap\n(per Participant)', fontsize=24)

sns.heatmap(P1_mat, ax=axs[0,0])
axs[0,0].set_title('P1', fontsize=18)

sns.heatmap(P2_mat, ax=axs[0,1])
axs[0,1].set_title('P2', fontsize=18)

sns.heatmap(P3_mat, ax=axs[1,0])
axs[1,0].set_title('P3', fontsize=18)

sns.heatmap(P4_mat, ax=axs[1,1])
axs[1,1].set_title('P4', fontsize=18)

fig.tight_layout()
plt.savefig('./plots/corr_part_heatmap.pdf')
plt.show()
```

# Correlation of task averages

```python
mean_path = './plot_data/task_means/'

task_corr = {'emg': joblib.load(mean_path+'emg'),
           'body': joblib.load(mean_path+'qom_body'),
           'head': joblib.load(mean_path+'qom_head'),
           'pupil': joblib.load(mean_path+'pupillometry'),
           'MIDI': joblib.load(mean_path+'midi'),
           'cogn_diff': joblib.load(mean_path+'cogn_diff'),
           'tech_diff': joblib.load(mean_path+'tech_diff')
          }

task_corr = pd.DataFrame(task_corr)
```

```python
task_corr
```

```python
task_mat = task_corr.corr()
round(task_corr,2)


# Plotting
fig, ax = plt.subplots(figsize=(7, 7))
fig.suptitle('Correlation Heatmap\n(Task Average)', fontsize=20)

sns.heatmap(task_mat)

fig.tight_layout()
plt.savefig('./plots/corr_task_heatmap.pdf')
plt.show()
```

```python
task_corr.corr()
```

```python
fig, ax = plt.subplots(figsize=(5,5))

sns.regplot(x='tech_diff', y='body', data = task_corr, label='P1-P4\n(mean)')

plt.title('Technical Difficulty and QoM (Body)\nCorrelation (Pearson: 0.791)', fontsize=16)

ax.set_ylabel('QoM (Body)')
ax.set_yticklabels([])
ax.set_xlabel('Technical Difficulty')
ax.set_xticklabels([])

fig.tight_layout()
plt.savefig('./plots/corr_best_mean.pdf')
plt.show()
```
